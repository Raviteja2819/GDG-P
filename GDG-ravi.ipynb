```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

import urllib3
import urllib
import json
import requests
urlLocation = 'http://api.open-notify.org/iss-now.json'
params = {'limit': 16, 'country': 'us', 'apikey': 'API-KEY'}
urlAstro = 'http://api.open-notify.org/astros.json'

response = urllib.request.urlopen(urlAstro)
result = json.loads(response.read())

response = requests.get(urlLocation, params=params)
long=float(response.json()['iss_position']['longitude'])
lat=float(response.json()['iss_position']['latitude'])

print('People in Space: ', result['number'])

people = result['people']
s="The people in ISS are : "
for p in people:
  s=s+p['name']+'\n'

print("Longitude is : "+str(long))
print("Latitude is : "+str(lat))
```

```python
fig = plt.subplots(figsize=(16, 16))
m = Basemap(projection='hammer',lon_0=0)
m.drawmapboundary(fill_color='#99ffff')
m.fillcontinents(color='#cc9966',lake_color='#99ffff')
from mpl_toolkits.basemap import Basemap

x, y = m(long, lat)
    
plt.plot(x, y, 'ok', markersize=10)
plt.text(x, y, ' ISS', fontsize=15);
plt.text(-175, +68, s,fontsize=15,fontweight='light',
                    ha='left',va='center',color='k')
```
